```python
import urllib2
from bs4 import BeautifulSoup # Parsing library
import pandas as pd
```

```python
source = urllib2.urlopen('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table')
source # It is a socket: similar to opening a file
soup = BeautifulSoup(source) 
print (soup)
```

```python
# Use BeautifulSoups Method .prettify( ) to show HTML  Document as a nested data structure
print soup.prettify( )
```

```python
table = soup.find("table", {"class" : "wikitable sortable"})
print table
```

```python
#There are 4 tables on the webpage. The table I am interested in has 7 columns and other tables have less than 7 columns. This can be verified by iterating over all the <tr HTML tags and printing out row lengths.

for i, row in enumerate(soup.findAll("tr")):
    cells = row.findAll("td")
    print i, len(cells)
```

```python
Team_IOC = [ ]
No_of_SummerGames_Participation = [ ]
No_of_SummerGames_Won_GoldMedal = [ ]
No_of_SummerGames_Won_SilverMedal = [ ]
No_of_SummerGames_Won_BronzeMedal = [ ]
Total_No_of_MedalsWon_Summer = [ ]
No_of_WinterGames_Participation = [ ]
No_of_WinterGames_Won_GoldMedal = [ ]
No_of_WinterGames_Won_SilverMedal = [ ]
No_of_WinterGames_Won_BronzeMedal = [ ]
Total_No_of_MedalsWon_Winter = [ ]
CombinedTotal_No_of_Games_Participation = [ ]
CombinedTotal_No_of_Won_GoldMedal = [ ]
CombinedTotal_No_of_Won_SilverMedal = [ ]
CombinedTotal_No_of_Won_BronzeMedal = [ ]
CombinedTotal_No_of_All_Medals_Won = [ ]


# Iterate over all the <tr tags and append data to the empty lists
for row in soup.findAll("tr"):
    cells = row.findAll("td")
    if len(cells) == 16:
        Team_IOC.append(cells[0].findAll(text = True))
        No_of_SummerGames_Participation.append(cells[1].findAll(text = True))
        No_of_SummerGames_Won_GoldMedal.append(cells[2].findAll(text = True))
        No_of_SummerGames_Won_SilverMedal.append(cells[3].findAll(text = True))
        No_of_SummerGames_Won_BronzeMedal.append(cells[4].findAll(text = True))
        Total_No_of_MedalsWon_Summer.append(cells[5].findAll(text = True))
        
        No_of_WinterGames_Participation.append(cells[6].findAll(text = True))
        No_of_WinterGames_Won_GoldMedal.append(cells[7].findAll(text = True))
        No_of_WinterGames_Won_SilverMedal.append(cells[8].findAll(text = True))
        No_of_WinterGames_Won_BronzeMedal.append(cells[9].findAll(text = True))
        Total_No_of_MedalsWon_Winter.append(cells[10].findAll(text = True))
        
        CombinedTotal_No_of_Games_Participation.append(cells[11].findAll(text = True))
        CombinedTotal_No_of_Won_GoldMedal.append(cells[12].findAll(text = True))
        CombinedTotal_No_of_Won_SilverMedal.append(cells[13].findAll(text = True))
        CombinedTotal_No_of_Won_BronzeMedal.append(cells[14].findAll(text = True))
        CombinedTotal_No_of_All_Medals_Won.append(cells[15].findAll(text = True))
        
        
        
```

```python
Team_IOCN = [ ]
for i, c in enumerate(Team_IOC):
    Team_IOCN.append(Team_IOC[i][1])
    
Sum_Olymp_Game_Participation =[ ]
for i, c in enumerate(No_of_SummerGames_Participation):
    Sum_Olymp_Game_Participation.append(No_of_SummerGames_Participation[i][0])

Sum_Won_GoldMedals =[ ]
for i, c in enumerate(No_of_SummerGames_Won_GoldMedal):
    Sum_Won_GoldMedals.append(No_of_SummerGames_Won_GoldMedal[i][0])

Sum_Won_SilverMedals =[ ]
for i, c in enumerate(No_of_SummerGames_Won_SilverMedal):
    Sum_Won_SilverMedals.append(No_of_SummerGames_Won_SilverMedal[i][0])

Sum_Won_BronzeMedals =[ ]
for i, c in enumerate(No_of_SummerGames_Won_GoldMedal):
    Sum_Won_BronzeMedals.append(No_of_SummerGames_Won_BronzeMedal[i][0])

Sum_Won_Total_No_of_Medals =[ ]
for i, c in enumerate(Total_No_of_MedalsWon_Summer):
    Sum_Won_Total_No_of_Medals.append(Total_No_of_MedalsWon_Summer[i][0])
    
# WINter gaMES TABLE
Winter_Olymp_Game_Participation =[ ]
for i, c in enumerate(No_of_WinterGames_Participation):
    Winter_Olymp_Game_Participation.append(No_of_WinterGames_Participation[i][0])
    

Winter_Won_GoldMedals =[ ]
for i, c in enumerate(No_of_WinterGames_Won_GoldMedal):
    Winter_Won_GoldMedals.append(No_of_WinterGames_Won_GoldMedal[i][0])

Winter_Won_SilverMedals =[ ]
for i, c in enumerate(No_of_WinterGames_Won_SilverMedal):
    Winter_Won_SilverMedals.append(No_of_WinterGames_Won_SilverMedal[i][0])

Winter_Won_BronzeMedals =[ ]
for i, c in enumerate(No_of_WinterGames_Won_BronzeMedal):
    Winter_Won_BronzeMedals.append(No_of_WinterGames_Won_BronzeMedal[i][0])

Winter_Won_Total_No_of_Medals =[ ]
for i, c in enumerate(Total_No_of_MedalsWon_Winter):
    Winter_Won_Total_No_of_Medals.append(Total_No_of_MedalsWon_Winter[i][0])
    
    
# COMIBINED TOTAL Olympic Games:SUMMER + WINTER

CombinedTotal_Olymp_Game_Participation =[ ]
for i, c in enumerate(CombinedTotal_No_of_Games_Participation):
    CombinedTotal_Olymp_Game_Participation.append(CombinedTotal_No_of_Games_Participation[i][0])
    
CombinedTotal_Won_GoldMedals =[ ]
for i, c in enumerate(CombinedTotal_No_of_Won_GoldMedal):
    CombinedTotal_Won_GoldMedals.append(CombinedTotal_No_of_Won_GoldMedal[i][0])

CombinedTotal_Won_SilverMedals =[ ]
for i, c in enumerate(CombinedTotal_No_of_Won_SilverMedal):
    CombinedTotal_Won_SilverMedals.append(CombinedTotal_No_of_Won_SilverMedal[i][0])

CombinedTotal_Won_BronzeMedals =[ ]
for i, c in enumerate(CombinedTotal_No_of_Won_BronzeMedal):
    CombinedTotal_Won_BronzeMedals.append(CombinedTotal_No_of_Won_BronzeMedal[i][0])

CombinedTotal_Won_Total_No_of_Medals =[ ]
for i, c in enumerate(CombinedTotal_No_of_All_Medals_Won):
    CombinedTotal_Won_Total_No_of_Medals.append(CombinedTotal_No_of_All_Medals_Won[i][0])  
    
print Team_IOCN[:5]
```

```python
# Creating a Pandas DataFrame
SummerOlympicGames = pd.DataFrame(data=[Team_IOCN,Sum_Olymp_Game_Participation,Sum_Won_GoldMedals,Sum_Won_SilverMedals,
                                        Sum_Won_BronzeMedals,Sum_Won_Total_No_of_Medals]).transpose()
```
